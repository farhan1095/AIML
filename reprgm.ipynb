```python
import pandas as pd
import numpy as np
sr1 = pd.Series([1, 2, 3])
sr2 = pd.Series([2, 3, 6])
print("Original Series:")
print("sr1:")
print(sr1)
print("sr2:")
print(sr2)
print("\nItems of a given series not present in another given series:")
sr11 = pd.Series(np.union1d(sr1, sr2))
sr22 = pd.Series(np.intersect1d(sr1, sr2))
result = sr11[~sr11.isin(sr22)]
print(result)
```

```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Salary': [50000, 60000, 55000],
    'Age': [25, 32, 28]
}
df = pd.DataFrame(data)
df['Updated_Salary'] = df['Salary'].apply(lambda x: x * 1.10)
df['Age_Group'] = df['Age'].apply(lambda x: 'Young' if x < 30 else 'Adult')
print("Updated DataFrame:\n", df)
```

```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Salary': [50000, 60000, 55000],
    'Age': [25, 32, 28]
}
df = pd.DataFrame(data)
df['Name_Upper'] = df['Name'].map(lambda x: x.upper())
salary_levels = {50000: 'Low', 55000: 'Medium', 60000: 'High'}
df['Salary_Level'] = df['Salary'].map(salary_levels)
print("Updated DataFrame:\n", df)
```

```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Aarav'],
    'Salary': [50000, 60000, 55000, 53000],
    'Age': [25, 32, 28, 24]
}
df = pd.DataFrame(data)
filtered_salary = df[df['Salary'] > 52000]
filtered_name = df[df['Name'].apply(lambda x: x.startswith('A'))]
print("Employees with salary > 52000:\n", filtered_salary)
print("\nEmployees whose names start with 'A':\n", filtered_name)
```

```python
import pandas as pd
from functools import reduce
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 32, 28]
}
df = pd.DataFrame(data)
df['Name_Age'] = df.apply(
    lambda row: reduce(lambda a, b: str(a) + '_' + str(b), [row['Name'], row['Age']]),
    axis=1
)
print("Updated DataFrame:\n", df)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
np.random.seed(42)
date_rng = pd.date_range(start='2022-01-01', end='2023-12-31', freq='D')
seasonality = np.sin(2 * np.pi * date_rng.dayofyear / 365)
noise = np.random.randn(len(date_rng))
time_series_data = noise + 0.2 * seasonality
time_series_df = pd.DataFrame({'Date': date_rng, 'Value': time_series_data})
time_series_df.set_index('Date', inplace=True)
result = seasonal_decompose(time_series_df['Value'], model='additive', period=365)
decomp_df = pd.DataFrame({
    'Observed': result.observed,
    'Trend': result.trend,
    'Seasonal': result.seasonal,
    'Residual': result.resid
}).dropna()
fig, axs = plt.subplots(4, 1, figsize=(14, 10), sharex=True)

axs[0].plot(decomp_df.index, decomp_df['Observed'], label='Observed', color='blue')
axs[0].set_ylabel('Observed')
axs[0].legend()

axs[1].plot(decomp_df.index, decomp_df['Trend'], label='Trend', color='orange')
axs[1].set_ylabel('Trend')
axs[1].legend()

axs[2].plot(decomp_df.index, decomp_df['Seasonal'], label='Seasonal', color='green')
axs[2].set_ylabel('Seasonal')
axs[2].legend()

axs[3].plot(decomp_df.index, decomp_df['Residual'], label='Residual', color='red')
axs[3].set_ylabel('Residual')
axs[3].set_xlabel('Date')
axs[3].legend()

plt.suptitle('Seasonal Decomposition of Time Series Data', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = {'CarType': ['Sedan', 'SUV', 'Truck', 'Hatchback', 'Sedan'],
'Frequency': [25, 15, 10, 20, 30]}
df = pd.DataFrame(data)
plt.figure(figsize=(8, 6))
plt.bar(df['CarType'], df['Frequency'], color='skyblue')
plt.title('Distribution of Car Types')
plt.xlabel('Car Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr'],
'Product_A': [100, 120, 80, 90],
'Product_B': [80, 90, 110, 100],
'Product_C': [70, 95, 85, 75]}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
bar_width = 0.2
index = df.index
plt.bar(index - bar_width, df['Product_A'], bar_width, label='Product A')
plt.bar(index, df['Product_B'], bar_width, label='Product B')
plt.bar(index + bar_width, df['Product_C'], bar_width, label='Product C')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Comparison')
plt.xticks(index, df['Month'])
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
categories = ['Category A', 'Category B', 'Category C']
subcategories = ['Subcategory 1', 'Subcategory 2', 'Subcategory 3']
proportions = [[0.2, 0.1, 0.1], [0.1, 0.2, 0.0], [0.1, 0.0, 0.2]]
plt.figure(figsize=(10, 6))
bottom = [0] * len(subcategories)
colors = ['lightcoral', 'lightskyblue', 'lightgreen']

for i, category in enumerate(categories):
    plt.bar(subcategories, proportions[i], label=category, bottom=bottom, color=colors[i])
    bottom = [sum(x) for x in zip(bottom, proportions[i])]
plt.title('Proportions in Subcategories by Category')
plt.xlabel('Subcategory')
plt.ylabel('Proportion')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03'],
'Category_A': [0.4, 0.3, 0.5],
'Category_B': [0.3, 0.4, 0.2],
'Category_C': [0.3, 0.3, 0.3]}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
plt.figure(figsize=(10, 6))
plt.stackplot(df['Date'], df['Category_A'], df['Category_B'], df['Category_C'], labels=['Category A',
'Category B', 'Category C'], alpha=0.7)
plt.title('Stacked Area Chart of Proportions Over Time')
plt.xlabel('Date')
plt.ylabel('Proportion')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
